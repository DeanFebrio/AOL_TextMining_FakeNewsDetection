```python
import pandas as pd
```

### Merge Data Train

```python
train_bodies_path = "./data/train/train_bodies.csv"
train_stances_path = "./data/train/train_stances.csv"
```

```python
train_bodies = pd.read_csv(train_bodies_path)
train_bodies.head()
```

```python
train_stances = pd.read_csv(train_stances_path)
train_stances.head()
```

```python
data_train = pd.merge(train_bodies, train_stances, on="Body ID")
data_train.head()
```

```python
train_path = "./data/train/train.csv"
```

```python
data_train.to_csv(train_path, index=False)
```
