# Exploratory Data Analysis (EDA) using ydata-profiling

Notebook ini bertujuan untuk menghasilkan laporan EDA dari dataset secara otomatis.

```python
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Load dataset
# Defaulting to data/train/train.csv. Silahkan ganti path ini jika menggunakan dataset lain.
df = pd.read_csv('data/train/train.csv')
```

```python
# Generate Profile Report
# explorative=True adds additional text analysis and features suited for text data
profile = ProfileReport(df, title='Fake News Detection EDA', explorative=True)
```

```python
# Export report to HTML
# Hasil akan disimpan dengan nama eda_report.html
profile.to_file('eda_report.html')
```
