```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train_path = "./data/processed/train_processed.csv"
test_path = "./data/processed/test_processed.csv"
```

```python
train_data = pd.read_csv(train_path)
train_data.head()
```

```python
test_data = pd.read_csv(test_path)
test_data.head()
```

```python
df_train_split, df_val_split = train_test_split(
    train_data,
    test_size=0.1,
    random_state=42,
    stratify=train_data["Stance"],
)

df_train_split = df_train_split.reset_index(drop=True)
df_val_split = df_val_split.reset_index(drop=True)

print(f"Train split size: {len(df_train_split)}")
print(f"Validation split size: {len(df_val_split)}")

print("\n=== Train Split Distribution ===")
print(df_train_split["Stance"].value_counts().sort_index())

print("\n=== Validation Split Distribution ===")
print(df_val_split["Stance"].value_counts().sort_index())
```

```python
train_body_ids = set(df_train_split["Body ID"])
val_body_ids = set(df_val_split["Body ID"])

overlap = train_body_ids.intersection(val_body_ids)

print('Number of unique Body IDs in train:', len(train_body_ids))
print('Number of unique Body IDs in val:', len(val_body_ids))
print('Number of overlapping Body IDs:', len(overlap))
print('Percentage of val Body IDs in train: {:.2f}%'.format(len(overlap) / len(val_body_ids) * 100))
```
